4. Удалите признаки Memory и ScreenResolution

```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', sep=',', encoding='windows-1251')
df.head()
```

#### 1. Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x:x.split()[0])
df.head()
```

#### 2. Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory_amount'] = df['Memory'].apply(lambda x:x.split()[0][:-2])
df.head()
```

#### 3. Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
# df['Memory_type'] = df['Memory'].apply(lambda x:x.split()[1:])
df['Memory_Type'] = df['Memory'].apply(lambda x: x.replace(' ', '').split('B')[-1])
df.head()
```

#### 4. Удалите признаки Memory и ScreenResolution

```python
df_drop = df.drop(columns=['Memory', 'ScreenResolution'])
df_drop.head()
```
